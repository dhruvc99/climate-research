```python
import xarray as xr
from ipynb.fs.full.PrepData import remove_seasonality
```

```python
t2m_files = ["/home/data/ERA5/day/t2m/t2m_{}.nc".format(i) for i in range(1979, 2020)]
t2m = xr.open_mfdataset(t2m_files).load()
t2m = t2m.sel(time=~((t2m.time.dt.month==2) & (t2m.time.dt.day == 29)))
t2m = t2m.t2m
```

```python
lats = t2m.lat.data
lons = t2m.lon.data
```

```python
for lat in lats:
    print(lat)
    for lon in lons:
        print('   ' + str(lon))
        t2m.sel({'lat':lat, 'lon':lon}).data  = remove_seasonality(t2m.sel({'lat':lat, 'lon':lon}).load())
```

```python
t2m.to_netcdf('/home/dhruv/data/deseasonalized/t2m.nc')
```

```python
shflx_files = ["/home/data/ERA5/day/shflx/shflx_{}.nc".format(i) for i in range(1979, 2020)]
shflx = xr.open_mfdataset(shflx_files)
shflx = shflx.shflx.sel(time=~((shflx.time.dt.month==2) & (shflx.time.dt.day == 29)))
```

```python
extent = [-11.25, 40, 35, 75]

all_lats = shflx.lat.values
all_lons = shflx.lon.values

extent_lats, extent_lons = [], []

for lat in all_lats:
    if lat >= 25 and lat <= 65:
        print(lat)
        extent_lats.append(lat)
        
for lon in all_lons:
    if lon >= (-130+360) and lon <= (-65+360):
        print(lon)
        extent_lons.append(lon)
```

```python
for lat in extent_lats:
    print(lat)
    for lon in extent_lons:
        print('   ' + str(lon))
        shflx.sel({'lat':lat, 'lon':lon}).data  = remove_seasonality(shflx.sel({'lat':lat, 'lon':lon}).load())
```

```python
shflx.to_netcdf('/home/dhruv/data/deseasonalized/europe/shflx.nc')
```
